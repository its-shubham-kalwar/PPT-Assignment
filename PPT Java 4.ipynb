### Q1.1. Write a program to show Interface Example in java?

````python
Certainly! Here's an example program in Java that demonstrates the usage of interfaces:

```java
// Interface representing a printable object
interface Printable {
    void print();
}

// Class implementing the Printable interface
class Book implements Printable {
    private String title;

    public Book(String title) {
        this.title = title;
    }

    @Override
    public void print() {
        System.out.println("Printing book: " + title);
    }
}

// Class implementing the Printable interface
class Magazine implements Printable {
    private String name;

    public Magazine(String name) {
        this.name = name;
    }

    @Override
    public void print() {
        System.out.println("Printing magazine: " + name);
    }
}

// Main class to run the program
public class InterfaceExample {
    public static void main(String[] args) {
        Printable printable1 = new Book("The Java Programming Language");
        Printable printable2 = new Magazine("Tech Today");

        // Call the print() method on printable objects
        printable1.print();
        printable2.print();
    }
}
```

Output:
```
Printing book: The Java Programming Language
Printing magazine: Tech Today
```

In this example, we have an interface called `Printable` that declares a single method `print()`.
The interface serves as a contract, defining a common behavior that classes implementing the 
interface must implement.

The `Book` class and the `Magazine` class both implement the `Printable` interface. They provide 
their own implementations of the `print()` method.

In the `main()` method, we create objects of `Book` and `Magazine` classes and assign them to 
variables of type `Printable`. This is possible because the classes implement the `Printable` 
interface. We can then call the `print()` method on these objects, which invokes the appropriate
implementation based on the actual object type.

The interface provides a way to achieve polymorphism, where objects of different classes can be 
treated interchangeably based on their common interface. It allows for loose coupling, modularity, 
and code reusability.
````

### Q2.Write a program a Program with 2 concrete method and 2 abstract method in java ?

````python
Certainly! Here's an example program in Java that includes a class with two concrete methods and
two abstract methods:

```java
// Abstract class with concrete and abstract methods
abstract class AbstractClass {
    // Concrete method
    public void concreteMethod1() {
        System.out.println("This is a concrete method 1.");
    }

    // Concrete method
    public void concreteMethod2() {
        System.out.println("This is a concrete method 2.");
    }

    // Abstract method 1 (to be implemented by subclasses)
    public abstract void abstractMethod1();

    // Abstract method 2 (to be implemented by subclasses)
    public abstract void abstractMethod2();
}

// Concrete subclass of AbstractClass
class ConcreteClass extends AbstractClass {
    // Implementation of abstractMethod1
    public void abstractMethod1() {
        System.out.println("Implementation of abstractMethod1.");
    }

    // Implementation of abstractMethod2
    public void abstractMethod2() {
        System.out.println("Implementation of abstractMethod2.");
    }
}

// Main class to run the program
public class AbstractExample {
    public static void main(String[] args) {
        ConcreteClass concreteObj = new ConcreteClass();

        // Call concrete methods
        concreteObj.concreteMethod1();
        concreteObj.concreteMethod2();

        // Call abstract methods
        concreteObj.abstractMethod1();
        concreteObj.abstractMethod2();
    }
}
```

Output:
```
This is a concrete method 1.
This is a concrete method 2.
Implementation of abstractMethod1.
Implementation of abstractMethod2.
```

In this example, we have an abstract class `AbstractClass` with two concrete methods 
(`concreteMethod1` and `concreteMethod2`) and two abstract methods (`abstractMethod1` and
`abstractMethod2`).

The `ConcreteClass` is a concrete subclass of `AbstractClass` and provides implementations
for the abstract methods. It inherits the concrete methods from the abstract class.

In the `main()` method, we create an object of the `ConcreteClass` and call both the concrete 
and abstract methods. The concrete methods are executed as defined in the abstract class, 
while the abstract methods are implemented in the `ConcreteClass` and provide specific functionality.

Note: Abstract methods do not have a method body and are declared with the `abstract` keyword. 
Concrete classes that extend an abstract class must implement all abstract methods defined in 
the abstract class.
````

### Q3.Write a program  to show the use of functional interface in java?

````python
Certainly! Here's an example program in Java that demonstrates the use of a functional interface 
and lambda expressions:

```java
// Functional interface with a single abstract method
@FunctionalInterface
interface MathOperation {
    int performOperation(int a, int b);
}

// Main class to run the program
public class FunctionalInterfaceExample {
    public static void main(String[] args) {
        // Lambda expression for addition operation
        MathOperation addition = (a, b) -> a + b;
        int result1 = addition.performOperation(5, 3);
        System.out.println("Addition Result: " + result1);

        // Lambda expression for subtraction operation
        MathOperation subtraction = (a, b) -> a - b;
        int result2 = subtraction.performOperation(8, 4);
        System.out.println("Subtraction Result: " + result2);
    }
}
```

Output:
```
Addition Result: 8
Subtraction Result: 4
```

In this example, we have a functional interface called `MathOperation` with a single abstract
method `performOperation()`. The `@FunctionalInterface` annotation is used to indicate that this
interface is intended to be used as a functional interface, which means it can be used with
lambda expressions.

In the `main()` method, we create lambda expressions for addition and subtraction operations 
using the `MathOperation` interface. The lambda expressions define the behavior of the abstract
method `performOperation()` by performing the respective mathematical operations.

We then call the `performOperation()` method on the lambda expressions, passing the operands to
get the results. The results are printed to the console.

Functional interfaces and lambda expressions allow us to write more concise and expressive code.
They facilitate the implementation of functional programming concepts and enable the use of lambda
expressions as first-class citizens in Java.
````

### Q4.What is an interface in Java?

````python
In Java, an interface is a reference type that defines a contract or set of methods that a 
class must implement. It provides a way to achieve abstraction and define a common behavior
that multiple classes can adhere to. Here are some key points about interfaces in Java:

1. Definition: An interface is declared using the `interface` keyword, followed by the interface
name. It can include abstract methods, default methods, static methods, and constant fields.

2. Abstraction: An interface represents an abstraction of behavior or functionality. It defines 
a set of methods that implementing classes must provide, without specifying the implementation 
details. It serves as a contract that guarantees certain behavior.

3. Multiple Inheritance: Unlike classes, Java supports multiple inheritance of interfaces. 
A class can implement multiple interfaces, allowing it to inherit and provide the behavior 
defined by those interfaces.

4. Implementation: Classes that implement an interface must provide the implementation for 
all the abstract methods declared in the interface. They do this by using the `implements` 
keyword followed by the interface name in the class declaration.

5. Method Signatures: In an interface, methods are declared without a body (except for default 
and static methods). They provide the method signature (name, return type, parameters), but the
implementation is left to the implementing classes.

6. Default and Static Methods: Java 8 introduced default and static methods in interfaces. 
Default methods have a default implementation, which can be overridden by implementing classes. 
Static methods in interfaces are associated with the interface itself and can be invoked using 
the interface name.

7. Constants: Interfaces can include constant fields, which are implicitly `public`, `static`, and `final`.
These fields can be accessed using the interface name followed by the field name.

Interfaces provide a powerful mechanism for achieving abstraction, polymorphism, and code 
reusability in Java. They allow for loose coupling, modular design, and the ability to define 
contracts that ensure consistent behavior across multiple classes. They are widely used in Java 
programming, especially in scenarios where multiple classes need to adhere to a common set of 
methods or behaviors.
````

### Q6.What is the lambda expression of Java 8?

````python
In Java 8, lambda expressions were introduced as a new feature to facilitate functional programming. 
A lambda expression is a concise way to represent a method as an expression. It allows you to pass 
behavior as a parameter to methods, define inline functions, and enable functional programming constructs.

The syntax of a lambda expression consists of the following parts:

```
(parameters) -> expression
```

or

```
(parameters) -> { statements }
```

Here's a breakdown of each part:

- Parameters: The parameters represent the inputs to the lambda expression. They can be zero or 
more parameters, enclosed in parentheses. If there is only one parameter, the parentheses can be omitted.
- Arrow Operator: The arrow operator (`->`) separates the parameter list from the body of the 
lambda expression.
- Expression or Statements: The expression or statements represent the behavior of the lambda expression.
For a single expression, it can be specified directly. For multiple statements, they are enclosed in 
curly braces.

Lambda expressions can be used in various contexts, including:

1. Functional Interfaces: Lambda expressions are commonly used with functional interfaces, which are 
interfaces with a single abstract method. Instead of implementing the interface with a separate class,
you can use a lambda expression to provide the implementation directly.

2. Stream API: The Stream API introduced in Java 8 allows for efficient processing of collections. 
Lambda expressions are frequently used to specify the behavior for filtering, mapping, reducing, 
and other operations on streams.

3. Concurrency and Parallelism: Lambda expressions facilitate the use of functional interfaces in 
concurrency and parallelism. They can be used with constructs such as `Runnable`, `Callable`, 
`CompletableFuture`, and functional interfaces in the `java.util.concurrent` package.

Lambda expressions in Java 8 help simplify the syntax for writing code in functional programming style,
improve code readability, enable concise and expressive code, and support the use of higher-order functions.
They are a powerful addition to the Java language and have significantly enhanced the programming 
capabilities of Java developers.
````

### Q7.Can you pass lambda expressions to a method? When?

````python
Yes, in Java, you can pass lambda expressions as arguments to methods. This feature is made possible by the 
functional interfaces, which are interfaces that have a single abstract method. Since lambda expressions 
can be used to provide an implementation for the single abstract method of a functional interface, they
can be used as arguments wherever a functional interface is expected.

Lambda expressions are commonly used in scenarios such as:

1. Callbacks: Methods that expect callbacks can receive lambda expressions as arguments. For example, 
in event-driven programming or asynchronous operations, you can pass lambda expressions to handle events 
or execute tasks upon completion.

2. Higher-Order Functions: Methods that operate on or return other functions can take lambda expressions 
as parameters. This enables the implementation of higher-order functions, which are functions that accept 
or return other functions.

3. Functional Operations: Methods that perform functional operations on collections or streams, such as 
filtering, mapping, sorting, or reducing, can receive lambda expressions as arguments to define the
desired behavior.

Here's an example to illustrate passing lambda expressions to a method:

```java
import java.util.Arrays;
import java.util.List;

interface StringOperation {
    String performOperation(String s);
}

public class LambdaAsMethodArgumentExample {
    public static void processStrings(List<String> strings, StringOperation operation) {
        for (String s : strings) {
            String result = operation.performOperation(s);
            System.out.println(result);
        }
    }

    public static void main(String[] args) {
        List<String> names = Arrays.asList("Alice", "Bob", "Charlie");

        // Pass a lambda expression as an argument
        processStrings(names, (s) -> "Hello, " + s + "!");

        // Pass a different lambda expression
        processStrings(names, (s) -> s.toUpperCase());
    }
}
```

Output:
```
Hello, Alice!
Hello, Bob!
Hello, Charlie!
ALICE
BOB
CHARLIE
```

In this example, we have a functional interface `StringOperation` with a single abstract method 
`performOperation()`. The `processStrings()` method takes a list of strings and a `StringOperation`
as arguments. It applies the `StringOperation` on each string in the list and performs some operation
defined by the lambda expression.

In the `main()` method, we call the `processStrings()` method with different lambda expressions as arguments.
These lambda expressions define the desired behavior for transforming or processing the strings.

By passing lambda expressions as arguments, you can provide dynamic behavior to methods, allowing for greater
flexibility, code reuse, and readability.
````

### Q8.What is the functional interface in Java 8?

````python
In Java 8, a functional interface is an interface that contains only one abstract method. It is also known
as a Single Abstract Method (SAM) interface or an interface with a functional method. The concept of 
functional interfaces is closely related to the functional programming paradigm and enables the use of 
lambda expressions and method references.

Java 8 introduced the `@FunctionalInterface` annotation to indicate that an interface is intended to be 
a functional interface. While it is not strictly required to use this annotation, it serves as a 
documentation tool and allows the compiler to validate that the interface has only one abstract method.

Functional interfaces provide the foundation for lambda expressions and method references in Java 8.
Lambda expressions can be used to provide an implementation for the single abstract method of a functional
interface, allowing for concise and expressive code.

Here's an example of a functional interface:

```java
@FunctionalInterface
interface MathOperation {
    int performOperation(int a, int b);
}
```

In this example, `MathOperation` is a functional interface because it contains a single abstract method 
`performOperation()`. The `@FunctionalInterface` annotation indicates that this interface is intended to 
be used as a functional interface, although it is not strictly necessary.

Functional interfaces are commonly used in scenarios where behavior needs to be passed as a parameter, such
as callbacks, stream operations, and concurrency. They provide a way to write more concise and expressive code,
enable functional programming constructs, and promote code modularity and reusability.

Java 8 also introduced several functional interfaces in the `java.util.function` package, such as `Predicate`,
`Function`, `Consumer`, and `Supplier`. These interfaces provide common functional operations and can be used 
directly or as a basis for creating custom functional interfaces.

Functional interfaces and lambda expressions are powerful features in Java 8 that enhance the 
expressiveness and flexibility of the language, enabling functional programming constructs and
more concise coding patterns.
````
