### Q1.Write a simple Banking System program by using OOPs concept where you can get account Holder name balance etc?

````python
Certainly! Here's an example of a simple banking system program in Java that demonstrates the
concept of object-oriented programming (OOP). The program allows you to create a bank account,
deposit and withdraw funds, and check the account balance:

```java
import java.util.Scanner;

// BankAccount class representing a bank account
class BankAccount {
    private String accountHolderName;
    private double balance;

    // Constructor
    public BankAccount(String accountHolderName, double initialBalance) {
        this.accountHolderName = accountHolderName;
        this.balance = initialBalance;
    }

    // Method to deposit funds into the account
    public void deposit(double amount) {
        balance += amount;
        System.out.println("Amount deposited successfully.");
    }

    // Method to withdraw funds from the account
    public void withdraw(double amount) {
        if (amount <= balance) {
            balance -= amount;
            System.out.println("Amount withdrawn successfully.");
        } else {
            System.out.println("Insufficient balance.");
        }
    }

    // Method to check the account balance
    public double getBalance() {
        return balance;
    }

    // Method to display account information
    public void displayAccountInfo() {
        System.out.println("Account Holder Name: " + accountHolderName);
        System.out.println("Account Balance: $" + balance);
    }
}

// Main class to run the program
public class BankingSystem {
    public static void main(String[] args) {
        // Create a scanner object for user input
        Scanner scanner = new Scanner(System.in);

        // Prompt the user to enter account details
        System.out.print("Enter Account Holder Name: ");
        String accountHolderName = scanner.nextLine();
        System.out.print("Enter Initial Balance: ");
        double initialBalance = scanner.nextDouble();

        // Create a bank account object
        BankAccount account = new BankAccount(accountHolderName, initialBalance);

        // Display account information
        account.displayAccountInfo();

        // Perform banking operations
        System.out.print("\nEnter amount to deposit: ");
        double depositAmount = scanner.nextDouble();
        account.deposit(depositAmount);

        System.out.print("\nEnter amount to withdraw: ");
        double withdrawAmount = scanner.nextDouble();
        account.withdraw(withdrawAmount);

        // Display updated account information
        account.displayAccountInfo();

        // Close the scanner
        scanner.close();
    }
}
```

In this program, the `BankAccount` class represents a bank account and provides methods for 
depositing, withdrawing, checking the balance, and displaying the account information. 
The `BankingSystem` class is the main class that creates a bank account object, interacts 
with the user to perform banking operations, and displays the account information.

You can compile and run the program to test it. It will prompt you to enter the account holder 
name, initial balance, deposit amount, and withdraw amount. The program will then perform the 
specified banking operations and display the updated account information.

Note: This is a basic example for illustrative purposes. In real-world scenarios, you would 
typically handle input validation, error handling, and additional features like multiple accounts,
interest calculations, and transaction history.
````

### Q2. Write a Program where you inherit method from parent class and show method Overridden Concept?

````python
Certainly! Here's an example program in Java that demonstrates method inheritance and method 
overriding concepts using a parent class and a child class:

```java
// Parent class
class Parent {
    public void display() {
        System.out.println("This is the parent class.");
    }
}

// Child class inheriting from the parent class
class Child extends Parent {
    @Override
    public void display() {
        System.out.println("This is the child class.");
    }
}

// Main class to run the program
public class InheritanceExample {
    public static void main(String[] args) {
        // Create objects of both parent and child classes
        Parent parentObj = new Parent();
        Child childObj = new Child();

        // Call the display() method on parent object
        parentObj.display();

        // Call the display() method on child object
        childObj.display();
    }
}
```

Output:
```
This is the parent class.
This is the child class.
```

In this program, we have a parent class `Parent` with a `display()` method. The child class 
`Child` extends the parent class, inheriting the `display()` method. The `Child` class overrides
the `display()` method using the `@Override` annotation to indicate that it is intentionally 
overriding the method from the parent class.

In the `main()` method, we create objects of both the parent and child classes. When we call 
the `display()` method on the parent object, it invokes the method defined in the parent class.
When we call the `display()` method on the child object, it invokes the overridden method 
defined in the child class.

This demonstrates the concept of method inheritance, where the child class inherits the method 
from its parent class. It also showcases method overriding, where the child class provides its 
own implementation of the method, which takes precedence over the parent class's implementation.

Note that the `@Override` annotation is not mandatory, but it helps in identifying and preventing
accidental errors when overriding methods. It ensures that the method signature in the child class
matches the method being overridden in the parent class.
````

### Q3.Write a program to show run time polymorphism in java?

````python
Certainly! Here's an example program in Java that demonstrates runtime polymorphism using 
inheritance and method overriding:

```java
// Parent class
class Animal {
    public void makeSound() {
        System.out.println("Animal makes a sound");
    }
}

// Child class 1
class Dog extends Animal {
    @Override
    public void makeSound() {
        System.out.println("Dog barks");
    }
}

// Child class 2
class Cat extends Animal {
    @Override
    public void makeSound() {
        System.out.println("Cat meows");
    }
}

// Main class to run the program
public class PolymorphismExample {
    public static void main(String[] args) {
        // Create objects of different types
        Animal animal1 = new Animal();
        Animal animal2 = new Dog();
        Animal animal3 = new Cat();

        // Call the makeSound() method on different objects
        animal1.makeSound();  // Invokes Animal's makeSound() method
        animal2.makeSound();  // Invokes Dog's makeSound() method
        animal3.makeSound();  // Invokes Cat's makeSound() method
    }
}
```

Output:
```
Animal makes a sound
Dog barks
Cat meows
```

In this program, we have a parent class `Animal` with a `makeSound()` method. Two child 
classes, `Dog` and `Cat`, extend the parent class and override the `makeSound()` method 
with their own implementations.

In the `main()` method, we create objects of different types: `Animal`, `Dog`, and `Cat`. 
We assign a `Dog` object to an `Animal` reference variable and a `Cat` object to another 
`Animal` reference variable. When we call the `makeSound()` method on these objects, the 
method invocation is resolved at runtime based on the actual type of the object.

Runtime polymorphism, also known as dynamic polymorphism, occurs when the method to be executed
is determined at runtime based on the actual object's type. In this example, the `makeSound()` 
method is overridden in the child classes, and the appropriate method is invoked based on the 
object's actual type during runtime.

Note that runtime polymorphism is achieved through method overriding and inheritance. By having 
a common method in the parent class and overriding it in the child classes, we can achieve
dynamic method dispatch, allowing the appropriate method implementation to be invoked based
on the actual object type at runtime.
````

### Q4.Write a program to show Compile time polymorphism in java?

````python
Certainly! Here's an example program in Java that demonstrates compile-time polymorphism using 
method overloading:

```java
public class PolymorphismExample {
    // Method with two integer parameters
    public static void add(int a, int b) {
        int sum = a + b;
        System.out.println("Sum of two integers: " + sum);
    }

    // Method with three integer parameters
    public static void add(int a, int b, int c) {
        int sum = a + b + c;
        System.out.println("Sum of three integers: " + sum);
    }

    // Method with two double parameters
    public static void add(double a, double b) {
        double sum = a + b;
        System.out.println("Sum of two doubles: " + sum);
    }

    public static void main(String[] args) {
        // Call the overloaded add() methods
        add(5, 10);
        add(2, 4, 6);
        add(3.5, 2.8);
    }
}
```

Output:
```
Sum of two integers: 15
Sum of three integers: 12
Sum of two doubles: 6.3
```

In this program, we have three overloaded versions of the `add()` method. Each method has a 
different number and/or type of parameters. This is an example of compile-time polymorphism,
also known as static polymorphism or method overloading.

In the `main()` method, we call the `add()` method with different arguments. The Java compiler 
determines which version of the `add()` method to invoke based on the number and types of the 
arguments provided. The decision is made at compile-time.

Compile-time polymorphism allows you to have multiple methods with the same name but different 
parameter lists. It enables you to perform different operations based on the number and types 
of the arguments passed to the method.

Note that method overloading is determined based on the method signature, which includes the method
name and the types and order of the parameters. The return type of the method is not considered 
for method overloading.
````

### Q5. Achieve loose coupling in java by using OOPs  concept?

````python
To achieve loose coupling in Java using object-oriented programming (OOP) concepts, you can follow
the principle of dependency inversion and apply techniques such as dependency injection and 
interface-based programming. Here's an explanation of how each concept helps in achieving loose coupling:

1. Dependency Inversion Principle (DIP):
   - The Dependency Inversion Principle suggests that high-level modules should not depend on 
low-level modules. Both should depend on abstractions.
   - By depending on abstractions rather than concrete implementations, you can achieve loose 
    coupling between classes.
   - This principle promotes the use of interfaces or abstract classes as the type of dependencies,
allowing flexibility in implementing different concrete classes without affecting the higher-level modules.
   - It enables you to change or extend implementations easily without modifying the dependent classes.

2. Dependency Injection (DI):
   - Dependency Injection is a design pattern that facilitates loose coupling by injecting dependencies 
into a class rather than having the class create its dependencies.
   - It involves providing dependencies to a class from an external source, such as a framework or a 
    container, rather than hard-coding them within the class itself.
   - By separating the creation and management of dependencies, you can easily replace or modify the 
dependencies without modifying the class that uses them.
   - Dependency injection can be achieved through constructor injection, setter injection, or method injection.

3. Interface-based Programming:
   - Interfaces define contracts or behaviors that classes can implement.
   - By programming to interfaces rather than concrete implementations, you achieve loose coupling 
    between components.
   - Classes can depend on interfaces, and different implementations of those interfaces can be used interchangeably.
   - This allows for easy substitution of implementations without affecting the code that depends on those interfaces.
   - Interface-based programming also promotes code reusability, modularity, and testability.

By applying these concepts together, you can achieve loose coupling in your Java code. This leads to 
code that is more modular, maintainable, and flexible, making it easier to manage changes and adapt
to evolving requirements.
````

### Q6. What is the benefit of encapsulation in java?

````python
Encapsulation is an important principle in object-oriented programming and refers to the bundling of 
data and methods within a class. It provides several benefits in Java:

1. Data Hiding: Encapsulation allows you to hide the internal details of an object and provide controlled
access to the data. By declaring variables as private and providing public getter and setter methods, 
you can control how the data is accessed and modified. This helps maintain the integrity and consistency
of the object's state.

2. Access Control: Encapsulation enables you to enforce access control on class members. By using access
modifiers (private, public, protected), you can restrict direct access to certain data and methods. 
This prevents unauthorized modifications and ensures that only the intended behavior is exposed to 
external code.

3. Code Organization and Maintainability: Encapsulation helps in organizing code by grouping related
data and methods together within a class. It improves code readability and makes it easier to understand
and maintain the codebase. Changes made to the internal implementation of a class do not impact the code
using the class, as long as the external interface (public methods) remains consistent.

4. Code Reusability: Encapsulation supports code reusability by providing well-defined interfaces. Once a
class is encapsulated, it can be easily used in other parts of the program or in other programs without 
worrying about its internal implementation. This promotes modularity and reduces code duplication.

5. Flexibility and Evolvability: Encapsulation allows you to modify the internal implementation of a class
without affecting the code that uses the class. As long as the public interface remains unchanged, other 
parts of the program can continue to use the class seamlessly. This makes it easier to enhance and extend
the functionality of existing code.

6. Security: Encapsulation provides a level of security by hiding sensitive data and exposing only the 
necessary operations through well-defined interfaces. It prevents unauthorized access and modification 
of internal data, thus enhancing the security of the program.

Encapsulation plays a crucial role in achieving the principles of abstraction, modularity, and
information hiding in Java. It promotes better design, code organization, and maintainability,
ultimately leading to more robust and scalable applications.
````

### Q7.Is java a t 100% Object oriented Programming language? If no why ?

````python
No, Java is not considered a 100% pure object-oriented programming language. While Java is 
predominantly object-oriented, there are a few factors that contribute to its classification 
as not being 100% pure object-oriented. Here's why:

1. Primitive Data Types: Java includes primitive data types such as int, char, boolean, etc., 
which are not objects. These types are not treated as objects and do not have associated methods
or inheritance like objects do.

2. Static Methods and Variables: Java allows the use of static methods and variables, which are
associated with the class itself rather than with specific instances of the class. Static methods
can be invoked without creating objects, and static variables are shared among all instances of a
class. This deviates from the pure object-oriented concept of encapsulating behavior within objects.

3. Wrapper Classes: Java provides wrapper classes (e.g., Integer, Double) to wrap primitive types
and treat them as objects. While this allows primitives to be used in object-oriented contexts, 
it still introduces a distinction between primitives and objects.

4. Procedural Programming Features: Java supports procedural programming constructs such as control
statements (if-else, loops) and procedural methodologies (top-down decomposition). Although these 
features are not required to use object-oriented programming, their inclusion in Java signifies a
departure from the strictest interpretation of pure object-oriented programming.

It's worth noting that Java is often considered a hybrid language, blending object-oriented and 
procedural programming paradigms. It heavily emphasizes object-oriented principles with its
class-based inheritance, encapsulation, and polymorphism. While it may not meet the criteria 
for being 100% pure object-oriented, Java remains widely used for object-oriented development 
and supports most of the essential features and benefits of object-oriented programming.
````

### Q8.What are the advantages of abstraction in java?

````python
Abstraction is a fundamental concept in Java and object-oriented programming. It involves hiding
unnecessary details and exposing only essential features of an object or system. Here are some 
advantages of abstraction in Java:

1. Simplifies Complex Systems: Abstraction allows you to represent complex systems or entities in a
simplified manner. By focusing on the essential features and hiding implementation details, you can
manage the complexity and make it easier to understand and work with the system.

2. Enhances Code Readability and Maintainability: Abstraction improves code readability by providing
a high-level view of a system. It allows developers to understand and work with the system without 
getting overwhelmed by the low-level details. It also improves code maintainability as changes to the
internal implementation can be isolated and hidden behind the abstract interfaces, minimizing the 
impact on other parts of the codebase.

3. Encourages Code Reusability: Abstraction promotes code reusability by providing well-defined interfaces.
When a class is abstracted, it can be easily used as a blueprint to create multiple derived classes with 
specific implementations. This saves development time and effort by avoiding code duplication and promotes
modular design.

4. Facilitates Loose Coupling: Abstraction helps in achieving loose coupling between components. 
By interacting with objects through abstract interfaces rather than concrete implementations, components
become decoupled and can be replaced or modified without affecting the rest of the system. This increases
flexibility, maintainability, and extensibility.

5. Supports Polymorphism: Abstraction is closely related to polymorphism, as it enables the use of 
abstract types and interfaces to represent a group of related objects. Polymorphism allows you to write 
code that can work with objects of different concrete classes that implement the same abstract interface.
This promotes flexibility and code reuse.

6. Improves Testability: Abstraction aids in writing unit tests and performing testing in isolation. 
By focusing on abstract interfaces and providing mock or stub implementations, you can test components 
independently without relying on the specific details of concrete implementations. This makes testing 
more manageable and effective.

Overall, abstraction in Java provides a way to simplify complex systems, enhance code readability and
maintainability, promote code reusability and loose coupling, support polymorphism, and improve testability.
It is a powerful tool for designing and building scalable, modular, and flexible applications.
````

### Q9.What is an abstraction explained with an Example?

````python
Abstraction is a fundamental concept in object-oriented programming that involves simplifying complex 
systems or entities by focusing on essential features and hiding unnecessary details. It allows you to 
create abstract representations of real-world objects or systems in order to model and work with them 
effectively. Here's an explanation of abstraction with an example:

Example: Shape Abstraction

Let's consider a simple example of representing shapes in a program. We can create an abstract class 
called `Shape` that defines common properties and behaviors of different shapes. The abstract class 
provides a blueprint for creating specific shape objects.

```java
// Abstract class representing a shape
abstract class Shape {
    protected String color;

    // Abstract method for calculating area
    public abstract double calculateArea();

    // Setter method for setting the color
    public void setColor(String color) {
        this.color = color;
    }

    // Getter method for getting the color
    public String getColor() {
        return color;
    }
}

// Concrete class representing a circle
class Circle extends Shape {
    private double radius;

    public Circle(double radius) {
        this.radius = radius;
    }

    // Implementation of calculateArea method for Circle
    public double calculateArea() {
        return Math.PI * radius * radius;
    }
}

// Concrete class representing a rectangle
class Rectangle extends Shape {
    private double length;
    private double width;

    public Rectangle(double length, double width) {
        this.length = length;
        this.width = width;
    }

    // Implementation of calculateArea method for Rectangle
    public double calculateArea() {
        return length * width;
    }
}

// Main class to run the program
public class AbstractionExample {
    public static void main(String[] args) {
        Circle circle = new Circle(5);
        circle.setColor("Red");

        Rectangle rectangle = new Rectangle(4, 6);
        rectangle.setColor("Blue");

        System.out.println("Circle Area: " + circle.calculateArea());
        System.out.println("Circle Color: " + circle.getColor());

        System.out.println("Rectangle Area: " + rectangle.calculateArea());
        System.out.println("Rectangle Color: " + rectangle.getColor());
    }
}
```

Output:
```
Circle Area: 78.53981633974483
Circle Color: Red
Rectangle Area: 24.0
Rectangle Color: Blue
```

In this example, the `Shape` class is an abstract class that represents the concept of a shape. 
It defines a protected `color` field and abstract methods for calculating the area of a shape.

The `Circle` and `Rectangle` classes extend the `Shape` class and provide concrete implementations 
for calculating the area. They also inherit the `color` field and the getter and setter methods from
the `Shape` class.

In the `main()` method, we create objects of the `Circle` and `Rectangle` classes, set their respective
colors, and calculate their areas. We can work with these objects using the common interface provided 
by the `Shape` class.

The abstraction in this example allows us to focus on the essential properties and behaviors of shapes
while hiding the specific implementation details. We can create various shapes by extending the `Shape`
class and provide their own implementations. This simplifies the representation and manipulation of shapes,
promotes code reuse, and makes the code more flexible and maintainable.
````

### Q10.What is the final class in Java?

````python
In Java, a final class is a class that cannot be extended or subclassed. When a class is declared as final,
it means that it cannot be inherited by other classes. Here are some key points about final classes:

1. Preventing Inheritance: By marking a class as final, you prevent other classes from inheriting or 
extending it. This is useful when you want to ensure that the class's implementation, behavior, or 
state remains unchanged and cannot be modified through inheritance.

2. Immutability: Final classes are often used in combination with immutable objects. By making a class
final, you ensure that its state cannot be modified by any subclass. This guarantees the immutability 
of objects and promotes thread-safety and consistency.

3. Security and Design Integrity: Final classes can provide security and protect sensitive operations 
or algorithms by preventing unauthorized modifications or extensions. Additionally, marking a class as
final can help maintain the design integrity of a framework or library by preventing unintentional 
changes to its core components.

4. Performance Optimization: Final classes can enable certain performance optimizations. Since a final
class cannot be subclassed, the JVM can make certain assumptions and perform optimizations at compile-time 
or runtime, resulting in potentially improved performance.

To declare a class as final in Java, you simply use the `final` keyword before the class declaration. 
Here's an example:

```java
final class FinalClass {
    // Class implementation
}
```

In this example, the `FinalClass` is marked as final, indicating that it cannot be subclassed. 
Any attempt to extend or subclass the `FinalClass` would result in a compilation error.

It's worth noting that while a final class cannot be subclassed, it can still use inheritance itself 
by extending other classes. Additionally, final methods and final variables can also be used within 
a final class.
````
